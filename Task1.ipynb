```python
import pandas as pd
df=pd.read_csv("Downloads/marketing_campaign.csv",sep="\t")
print("This is the Marketing campaign (Customer Personality Analysis) csv file:")
print(df.shape)
print(df.head())

print(df.isnull())

print("Before removing duplicate rows:", df.shape)
df = df.drop_duplicates()
print("After removing duplicate rows:", df.shape)

print("Standardize 'Education' column:")
df['Education'] = df['Education'].str.strip().str.title()

print("Converting Dt_Customer to datetime:")
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')

print("Renameing column headers to be clean and uniform:")
df.columns = df.columns.str.lower().str.replace(" ", "_")
print(df.columns)

print("Checking and fixing data types:")
df['year_birth'] = df['year_birth'].astype(int)
df['income'] = df['income'].astype(float)

df.to_csv("marketing_campaign_cleaned.csv", index=False)
print("FINAL DATASET VIEW.")
print(df.head())
```
